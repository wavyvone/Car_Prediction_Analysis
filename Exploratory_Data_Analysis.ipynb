# Imports

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data_URL = "./Data/CarPrice_Assignment.csv"

data = pd.read_csv(data_URL, index_col=0)

data.head()
```

# Price (Target Feature) Analysis

```python
fig = plt.figure()
price_data = data["price"]
sns.displot(price_data, kde=True)
plt.title("Car Price Distribution")

plt.show()

price_data.agg(["min", "max", "mean", "median", "std"]).to_frame().T
```

# Feature Analysis

```python
def feature_analysis_categorical(feature_name: str):
  plt.figure(figsize=(20,8))
  plt.subplot(1, 2, 1)
  data[feature_name].value_counts().plot(kind="barh")
  plt.title(f'Value Counts For {str(feature_name)}')
  plt.xlabel("Count")


  plt.subplot(1, 2, 2)
  sns.boxplot(y=feature_name, x="price", data=data)
  plt.title(f'Box Plot for {str(feature_name)} Vs Price')

  plt.show()

  feature_data = []
  for feature in data[feature_name].unique():
      feature_data.append([feature, data[data[feature_name] == feature]["price"].min(), \
        data[data[feature_name] == feature]["price"].max(), \
        data[data[feature_name] == feature]["price"].mean(), \
        data[data[feature_name] == feature]["price"].median(), \
        data[data[feature_name] == feature]["price"].std()])

  feature_data = pd.DataFrame(data=feature_data, columns=["feature", "min", "max", "mean", "median", "stdev"])
  feature_data.set_index("feature", inplace=True)
  print(feature_data.head())
```

```python
def feature_analysis_numerical(feature_name: str):
  plt.figure(figsize=(20,8))

  plt.subplot(1, 2, 1)
  plt.hist(data[feature_name], bins=20)
  plt.title(f'Histogram for {str(feature_name)}')
  plt.ylabel("Count")
  plt.xlabel(feature_name)

  plt.subplot(1, 2, 2)
  plt.scatter(data[feature_name], data["price"])
  plt.title(f'Scatter Plot for {str(feature_name)} Vs Price')
  plt.ylabel("Price")
  plt.xlabel(feature_name)

  plt.show()

  print(data[feature_name].describe())
```

```python
def two_feature_analysis_numerical(feature1: str, feature2):
    plt.scatter(data[feature1], data[feature2])
    plt.title(f'Scatter Plot for {str(feature1)} Vs {str(feature1)}')
    plt.ylabel(feature2)
    plt.xlabel(feature1)

    plt.show()
```

```python
sns.pairplot(data.select_dtypes(exclude="object"))
```

## Aspiration

```python
feature_analysis_categorical("aspiration")
```

The dataset contains many more standard aspiration cars than turbo aspiration cars.  Turbo cars have both higher median and mean prices.  However, as indicated by the overlap in box plot boxes, the prices are relatively similar.

There are many more price outliers for the standard aspiration cars, with a large number of cars in this category being higher in price.

## DoorNumber

```python
feature_analysis_categorical("doornumber")
```

Two- and four-door cars have both similar mean and median prices.  However, the mean and median are both slightly higher for four-door cars.  Both categories experience outliers, with a set of prices being larger.

## CarBody

```python
feature_analysis_categorical("carbody")
```

The majority of cars in the dataset are either sedans or hatchbacks.

Convertibles and hardtop cars experience a larger spread in prices, along with the largest median prices.  In contrast, the three remaining car body types have smaller median prices, as well as smaller variance in price.

Sedans had a larger number of price outliers, indicating that certain sedans sold for prices much higher than the median.

## DriveWheel

```python
feature_analysis_categorical("drivewheel")
```

Four-wheel drives sell for a higher price on average than front-wheel and rear-wheel drives. Four-wheel drives were sold at a larger range of prices (having a larger standard deviation) than the other two.

Front-wheel drives do have a few outliers.

## EngineLocation


```python
feature_analysis_categorical("enginelocation")
```

Cars with engines in the rear tend to be sold for higher prices than cars with engines in the front. However, there are a lot of outliers for cars with front engines and there is not too much data for cars with rear engines. 

## Wheel Base

```python
feature_analysis_numerical("wheelbase")
```

The average wheelbase size in our dataset is 98 inches, there are some evidence that shows the wheelbase increases the price increases. There are many outliers where the price is higher when the wheelbase is small. This could be due to the smaller wheelbase of compact expensive car

## CarLength

```python
feature_analysis_numerical("carlength")
```

The average car length is around 174 inches/5 yards, but as car length increases there is also a positive increase in price.

## CarWidth

```python
feature_analysis_numerical("carwidth")
```

The car width averages settle between the 65 inches, as the car widths increases the prices increases. More car materials are also required for car size increase which costs money.

```python
two_feature_analysis_numerical("carwidth", "carlength")
```

## CarHeight

```python
feature_analysis_numerical("carheight")
```

The average car height is approximately 54 inchese.

## CurbWeight

```python
feature_analysis_numerical("curbweight")
```

The average curb weight of the cars sold is about 2555 lbs. As car length increases, the price of the car generally increases as well. However, as curb weight increases, this relationship tends to be less present (there is more variation in price as curb weight increases). 

## EngineType

```python
feature_analysis_categorical("enginetype")
```

The vast majority of cars sold had an overhead camshaft (OHC) type of engine.  Cars of this engine type had an average price of approximately $11,574.  This engine type actually had the smallest mean and median prices of the different types, in addition to the most price outliers.

There are very few datapoints in the set for rotor and DOHCV type engines, therefore making these engine types more difficult to make conclusions about.

## CylinderNumber

```python
feature_analysis_categorical("cylindernumber")
```

On average, as the number of cylinders increases, the price also increases. However, on average cars with two-cylinder engines are more expensive than cars with four-cylinder engines.
