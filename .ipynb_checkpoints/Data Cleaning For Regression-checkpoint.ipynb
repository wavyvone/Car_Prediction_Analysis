# Import the data

```python
import pandas as pd
import numpy as np

data_URL = "./Data/CarPrice_Assignment.csv"

data = pd.read_csv(data_URL, index_col=0)

data.head()
```

# Transforming Categorical Features

```python
data.select_dtypes(include="object").columns
```

```python
data["fueltype"].unique()
data["fueltype"].value_counts()
data['fueltype_gas'] = np.where(data["fueltype"] == "gas", 1, 0)
data['fueltype_diesel'] = np.where(data["fueltype"] == "diesel", 1, 0)

data["aspiration"].unique()
data["aspiration"].value_counts()
data['aspiration_std'] = np.where(data["aspiration"] == "std", 1, 0)
data['aspiration_turbo'] = np.where(data["aspiration"] == "turbo", 1, 0)

data["doornumber"].unique()
data["doornumber"].value_counts()
data['doornumber_two'] = np.where(data["doornumber"] == "two", 1, 0)
data['doornumber_four'] = np.where(data["doornumber"] == "four", 1, 0)

data["carbody"].unique()
data["carbody"].value_counts()
data['carbody_sedan'] = np.where(data["carbody"] == "sedan", 1, 0)
data['carbody_hatchback'] = np.where(data["carbody"] == "hatchback", 1, 0)
data['carbody_wagon'] = np.where(data["carbody"] == "wagon", 1, 0)
data['carbody_hardtop'] = np.where(data["carbody"] == "hardtop", 1, 0)
data['carbody_convertible'] = np.where(data["carbody"] == "convertible", 1, 0)

data["drivewheel"].unique()
data["drivewheel"].value_counts()
data['drivewheel_fwd'] = np.where(data["drivewheel"] == "fwd", 1, 0)
data['drivewheel_rwd'] = np.where(data["drivewheel"] == "rwd", 1, 0)
data['drivewheel_4wd'] = np.where(data["drivewheel"] == "4wd", 1, 0)

data["enginelocation"].unique()
data["enginelocation"].value_counts()
data['enginelocation_front'] = np.where(data["enginelocation"] == "front", 1, 0)
data['enginelocation_rear'] = np.where(data["enginelocation"] == "rear", 1, 0)

data["enginetype"].unique()
data["enginetype"].value_counts()
data['enginetype_ohc'] = np.where(data["enginetype"] == "ohc", 1, 0)
data['enginetype_ohcf'] = np.where(data["enginetype"] == "ohcf", 1, 0)
data['enginetype_ohcv'] = np.where(data["enginetype"] == "ohcv", 1, 0)
data['enginetype_dohc'] = np.where(data["enginetype"] == "dohc", 1, 0)
data['enginetype_l'] = np.where(data["enginetype"] == "l", 1, 0)
data['enginetype_rotor'] = np.where(data["enginetype"] == "rotor", 1, 0)
data['enginetype_dohcv'] = np.where(data["enginetype"] == "dohcv", 1, 0)

data["cylindernumber"].unique()
data["cylindernumber"].value_counts()
data['cylindernumber_two'] = np.where(data["cylindernumber"] == "two", 1, 0)
data['cylindernumber_three'] = np.where(data["cylindernumber"] == "three", 1, 0)
data['cylindernumber_four'] = np.where(data["cylindernumber"] == "four", 1, 0)
data['cylindernumber_five'] = np.where(data["cylindernumber"] == "five", 1, 0)
data['cylindernumber_six'] = np.where(data["cylindernumber"] == "six", 1, 0)
data['cylindernumber_eight'] = np.where(data["cylindernumber"] == "eight", 1, 0)
data['cylindernumber_twelve'] = np.where(data["cylindernumber"] == "twelve", 1, 0)

data["fuelsystem"].unique()
data["fuelsystem"].value_counts()
data['fuelsystem_mpfi'] = np.where(data["fuelsystem"] == "mpfi", 1, 0)
data['fuelsystem_2bbl'] = np.where(data["fuelsystem"] == "2bbl", 1, 0)
data['fuelsystem_idi'] = np.where(data["fuelsystem"] == "idi", 1, 0)
data['fuelsystem_1bbl'] = np.where(data["fuelsystem"] == "1bbl", 1, 0)
data['fuelsystem_spdi'] = np.where(data["fuelsystem"] == "spdi", 1, 0)
data['fuelsystem_4bbl'] = np.where(data["fuelsystem"] == "4bbl", 1, 0)
data['fuelsystem_mfi'] = np.where(data["fuelsystem"] == "mfi", 1, 0)
data['fuelsystem_spfi'] = np.where(data["fuelsystem"] == "spfi", 1, 0)
```

```python
data = data.drop(columns=data.select_dtypes(include="object").columns)
```
